```python
# Check missing values
df.isnull().sum()
```

We can see that "Review Rating" has 37 missing values. We can replace them in a few ways, including either the mean or the median. Generally speaking, it's better to replace it with the median because the mean is affected by outliers, which can skew the distribution of the data.

However, just replacing the missing values with the cumulative review rating median is not good enough. This is because different categories of products might have different review rating systems. For example, ratings for clothing might differ from ratings for shoes; this can introduce bias into the dataset. To overcome this, we will be filling in missing values with the median for each respective category. 

```python
# Fill the missing values with the median for each respective category
# (e.g., missing clothes ratings get replaced with the median for only clothes, not anything else) 
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))

# Check to see if there are any missing values anymore
df.isnull().sum()
```

```python
# Have a standardized naming convention for the column names
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')

df.columns
```

```python
df = df.rename(columns=({'purchase_amount_(usd)': 'purchase_amount'}))  # Remove the '(usd)' to standardize name

df.columns
```

### Feature Engineering

```python
# Create a new column: age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)  # Split ages into 4 different-sized groups, based on data distribution
```

```python
df[['age', 'age_group']].head(20)
```

```python
# Create a new column: purchase_frequency_days

# Here, we are going to convert text frequencies into numerical values--makes it easier for analysis later on

# Mappings are done by days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(20)
```

```python
# Look at how the two following columns are related
df[['discount_applied', 'promo_code_used']].head(20)
```

These two columns are very much related. This is because if a promo code was used, then some sort of discount was applied to the purchase. Though sometimes, companies can apply discounts without promo codes (e.g., seasonal sales, exclusive discounts for members, etc.).

So the question is, *Do we need both columns?*

Delve more into the relationship between the two to see the answer...

```python
# See if both columns contain the same information
(df['discount_applied'] == df['promo_code_used']).all()
```

From the code above, we can see that both columns indeed carry the same info. This means one of the columns is redundant. In this case, let's remove `promo_code_used`.

```python
# Remove promo_code_used because it's redundant
df = df.drop('promo_code_used', axis=1)
df.columns
```
